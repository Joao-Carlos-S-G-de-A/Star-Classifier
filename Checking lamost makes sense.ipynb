# Checking lamost data makes sense

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def load_spectra_from_directory(directory, expected_shape=(5, 3748)):
    """Load all .npy files from a directory and return a list of numpy arrays with the expected shape."""
    # Initialize 5 lists to store each row of the spectra separately
    row_stacks = [[] for _ in range(5)]

    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".npy"):
                file_path = os.path.join(root, file)
                try:
                    spectra = np.load(file_path)
                    spectra = truncate_spectra(spectra, expected_shape[1])
                    if spectra.shape == expected_shape:
                        
                        print(f"Loaded {file} with shape {spectra.shape}")
                        
                        # Append each row of the spectra to its respective list
                        for i in range(5):
                            row_stacks[i].append(spectra[i])
                    else:
                        print(f"Skipping {file}: shape {spectra.shape} does not match expected {expected_shape}")
                except Exception as e:
                    print(f"Error loading {file}: {e}")

    # Convert each list to a NumPy array
    row_stacks = [np.stack(row_list) for row_list in row_stacks]
    return row_stacks

def truncate_spectra(spectra, num_elements):
    """Truncate the first num_elements of each spectra."""
    return spectra[:, :num_elements]


# Load galaxy spectra from the "training_npy/gal_spectra" directory
galaxy_spectra_dir = "training_npy/gal_spectra"
row_stacks = load_spectra_from_directory(galaxy_spectra_dir)


# Print the shape of each stacked row
for i, row_stack in enumerate(row_stacks):
    print(f"Shape of stacked row {i+1}: {row_stack.shape}")
```
